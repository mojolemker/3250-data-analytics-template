```python
# Imports
import pandas as pd
```

```python
# Load in CSV
song_characteristics_raw = pd.read_csv('/Users/joseflemker/Documents/GitHub/3250-data-analytics-template/data/raw/Spotify_Song_Attributes.csv', encoding='utf-8')
```

```python
# Examine first few rows and the data set as a whole
song_characteristics_raw.dtypes, print(song_characteristics_raw.head())
```

```python
# Check for exact duplicates
duplicates = song_characteristics_raw.duplicated(keep=False)
print(f"duplicate records: {len(song_characteristics_raw[song_characteristics_raw.duplicated(keep='first')])} \n")
```

```python
# Examine some of the first values and first duplicates
duplicates_df = song_characteristics_raw[song_characteristics_raw.duplicated(keep=False)]
print(duplicates_df.head())
```

```python
# Now the original
print(song_characteristics_raw.head())
```

Looks like all the data is duplicated twice so we will just drop the duplicates

```python
# Drop Duplicates
song_characteristics_df = song_characteristics_raw.drop_duplicates()
```

```python
# Double check there is no duplicates now
print(f"duplicate records: {len(song_characteristics_df[song_characteristics_df.duplicated(keep='first')])} \n")
```

```python
# Now checking for duplicate song titles
duped_songs = song_characteristics_df[song_characteristics_df['trackName'].duplicated(keep=False)]
print(display(duped_songs))
```

Looks like most of the songs just happen to have the same name so we need to also pull artistName

```python
# Now check with artist
duped_songs2 = song_characteristics_df[song_characteristics_df.duplicated(subset=['trackName', 'artistName'], keep=False)]
print(display(duped_songs2))
```

Duplicates are good now need to check for null values

```python
# Print the total missing data
print(f'Total missing data: {song_characteristics_df.isna().sum().sum()}')
# Print missing data per column
print(song_characteristics_df.isna().sum())
```

Seeing consisten 275 across all categories. Looks like we should drop all rows with danceability as null and verify the other columns go to 0

```python
# Drop rows wheer danceability is null
song_characteristics_df = song_characteristics_df[song_characteristics_df['danceability'].notna()]
```

```python
# Print missing data per column
print(song_characteristics_df.isna().sum())
```

This has cleared all but the genre column. Let's look at some of those values

```python
# Make new df for null genres
null_genres_df = song_characteristics_df[song_characteristics_df['genre'].isna()]
print(null_genres_df)
```

It appears to be a random assortment of no genres. Because it is a releatively small amount and is necessary for analysis we will just remove the rows where genre is null.

```python
# Drop rows wheer danceability is null
song_characteristics_df = song_characteristics_df[song_characteristics_df['genre'].notna()]
```

```python
# Print missing data per column
print(song_characteristics_df.isna().sum())
```

Now I need to ensure that the datatypes accurately represent the columns and we have only necessary columns included.

```python
# Observe data types
print(song_characteristics_df.dtypes)
```

```python
# Look at total uniques for each column to ensure we get al the categorical variables we need
print(song_characteristics_df.nunique())
```

```python
# Based off those numbers and column descriptions choose categorical variables
print(f'Total genres: {len(song_characteristics_df['genre'].unique())}')
print(f' Total keys: {len(song_characteristics_df['key'].unique())}')
print(f' Total time signature: {len(song_characteristics_df['time_signature'].unique())}')
print(f' Total modes: {len(song_characteristics_df['mode'].unique())}')
print(song_characteristics_df['genre'].unique())
print(song_characteristics_df['key'].unique())
print(song_characteristics_df['time_signature'].unique())
print(song_characteristics_df['mode'].unique())
```

```python
# Change columns to categorical
song_characteristics_df['genre_edited'] = song_characteristics_df['genre'].astype('category')
song_characteristics_df['key_edited'] = song_characteristics_df['key'].astype('category')
song_characteristics_df['time_signature_edited'] = song_characteristics_df['key'].astype('category')
song_characteristics_df['mode_edited'] = song_characteristics_df['mode'].astype('category')
```

```python
# Show new dtypes
print(song_characteristics_df.dtypes)
```

```python
# Select and order the correct columns into cleaned dataframe
clean_df_columns = ['trackName', 'artistName', 'genre_edited', 'danceability', 'energy', 'key_edited', 'loudness', 'mode_edited', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature_edited']
song_characteristics_df_clean = song_characteristics_df[clean_df_columns]
```

```python
# Rename to match data dictionairy
song_characteristics_df_clean = song_characteristics_df_clean.rename(columns={
                                     'trackName': 'track_name',
                                     'artistName': 'artist_name',
                                     'genre_edited': 'genre',
                                     'mode_edited': 'mode',
                                     'key_edited': 'key',
                                     'time_signature_edited': 'time_signature'
                                     })
```

```python
# Save the dataframe to a csv
song_characteristics_df_clean.to_csv('/Users/joseflemker/Documents/GitHub/3250-data-analytics-template/data/finished/kaggle_finished_data.csv', encoding='utf-8', index=False)
```
