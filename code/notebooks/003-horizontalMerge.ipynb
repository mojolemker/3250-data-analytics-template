# Merging Data files into one table

```python
# Imports
import pandas as pd
```

```python
# Load files in as dataframes
kaggle_df = pd.read_csv('/Users/joseflemker/Documents/GitHub/3250-data-analytics-template/data/finished/kaggle_finished_data.csv', encoding='utf-8')
kworb_df = pd.read_csv('/Users/joseflemker/Documents/GitHub/3250-data-analytics-template/data/finished/kworb_finished_data.csv', encoding='utf-8')
```

```python
# Merge the files on track name
merged_df = kworb_df.merge(kaggle_df, how='inner', on=["track_name", "artist_name"])
```

```python
# View new merged df
merged_df.to_csv('/Users/joseflemker/Documents/GitHub/3250-data-analytics-template/data/finished/merged_data_finished.csv', encoding = 'utf-8', index=False)
```
